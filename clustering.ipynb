```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
training_data = clustering_data_cleaned[['Avg_Rating', 'Rating_Count', 'Product_Price']]
training_data = scaler.fit_transform(training_data)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette_scores = []

# Loop to calculate inertia and silhouette score for different values of K
for k in range(2, 11):  # Start from 2 because silhouette score is not defined for K=1
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(training_data)  # Fit the model to the scaled data
    
    # Append inertia (within-cluster sum of squares) for the current K
    inertia.append(kmeans.inertia_)
    
    # Calculate and append silhouette score for the current K
    score = silhouette_score(training_data, kmeans.labels_)
    silhouette_scores.append(score)

# Create subplots to plot both Elbow and Silhouette Score graphs
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Elbow Method Plot
ax1.plot(range(2, 11), inertia, marker='o')
ax1.set_xlabel('Number of Clusters (K)')
ax1.set_ylabel('Inertia')
ax1.set_title('Elbow Method for Optimal K')

# Silhouette Score Plot
ax2.plot(range(2, 11), silhouette_scores, marker='o')
ax2.set_xlabel('Number of Clusters (K)')
ax2.set_ylabel('Silhouette Score')
ax2.set_title('Silhouette Score for Optimal K')

# Display the plots
plt.tight_layout()
plt.show()
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def initialize_centroids(X, K):
    random_sample_idxs = np.random.choice(X.shape[0], K, replace=False)
    centroids = X[random_sample_idxs]
    return centroids

# Function to assign samples to the nearest centroids
def create_clusters(X, centroids):
    clusters = [[] for _ in range(len(centroids))]
    for idx, sample in enumerate(X):
        centroid_idx = np.argmin([euclidean_distance(sample, point) for point in centroids])
        clusters[centroid_idx].append(idx)
    return clusters

# Function to calculate new centroids
def calculate_centroids(X, clusters):
    centroids = np.zeros((len(clusters), X.shape[1]))
    for idx, cluster in enumerate(clusters):
        cluster_points = X[cluster]
        centroids[idx] = np.mean(cluster_points, axis=0)
    return centroids

# Function to check for convergence
def is_converged(old_centroids, centroids):
    distances = [euclidean_distance(old, new) for old, new in zip(old_centroids, centroids)]
    return sum(distances) == 0

# Function to assign cluster labels to samples
def get_cluster_labels(clusters, n_samples):
    labels = np.empty(n_samples)
    for cluster_idx, cluster in enumerate(clusters):
        for sample_idx in cluster:
            labels[sample_idx] = cluster_idx
    return labels

# Main KMeans function
def kmeans(X, K, max_iters=100):
    
    centroids = initialize_centroids(X, K)
    
    for _ in range(max_iters):
        clusters = create_clusters(X, centroids)

        old_centroids = centroids
        centroids = calculate_centroids(X, clusters)

        if is_converged(old_centroids, centroids):
            break

    labels = get_cluster_labels(clusters, X.shape[0])
    return labels, centroids
```

```python
K_values = range(2, 11)  # Example: K from 2 to 10

silhouette_scores_manual = []

# Iterate over each value of K
for k in K_values:
    labels, centroids = kmeans(training_data, k)  # Run your custom kmeans function
    score = silhouette_score(training_data, labels)  # Calculate silhouette score
    silhouette_scores_manual.append(score)

# Plot the Silhouette Scores
plt.figure(figsize=(8, 6))
plt.plot(K_values, silhouette_scores_manual, marker='o')
plt.xticks(K_values)  # Show each K value on the x-axis
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Scores for Different K")
plt.show()
```
